## Base Scenario Audit

The pooling calculations are derived from the current-admissions occupancy audit using the appendix's partial-pooling probability logic.


```python
settings = SimulationSettings(run_length_days=365 * 5, warm_up_days=365 * 3, replications=30)
audit = run_replications(settings=settings)
pooling_table = build_pooling_results_table(audit)
pooling_table
```

## Table 3 Style Presentation

The table below mirrors the structure of the pooling results table in Monks et al.


```python
table_3_style = pooling_table.rename(
    columns={
        'dedicated_acute': 'Dedicated Acute',
        'dedicated_rehab': 'Dedicated Rehab',
        'pooled': 'Pooled',
        'acute_p_delay': 'P(delay) Acute',
        'rehab_p_delay': 'P(delay) Rehab',
        'acute_1_in_n': '1 in n Acute',
        'rehab_1_in_n': '1 in n Rehab',
    }
)
table_3_style
```

## Notes

- Complete pooling is estimated from the total audited occupancy across acute and rehab beds.
- Partial pooling uses the probability logic described in the supplementary appendix and an independence approximation between acute and rehab occupancies.
